```python
images_Arr[4795] = cv2.cvtColor(np.asarray(images_Arr[4795]),cv2.COLOR_GRAY2RGB)
twoDimage = np.asarray(images_Arr[4795]).reshape((-1,3))
twoDimage = np.float32(twoDimage)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 4
attempts=10
ret,label,center=cv2.kmeans(twoDimage,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((np.asarray(images_Arr[4795]).shape))
result_image
```

```python
#images_copy = images
kmeans_images = []
def kmeans_blur():
  ind=0
  for x in images_Arr: 
    if ind not in [1534,4042,4795,7798,8368]:
      twoDimage = np.asarray(x).reshape((-1,3))
      twoDimage = np.float32(twoDimage)
      criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
      K = 4
      attempts=10
      ret,label,center=cv2.kmeans(twoDimage,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
      center = np.uint8(center)
      res = center[label.flatten()]
      result_image = res.reshape((np.asarray(x).shape))
      kmeans_images.append(Image.fromarray(result_image.astype('uint8'), 'RGB'))
      print('Index Done ',ind)
    else:
      images_Arr[ind] = cv2.cvtColor(np.asarray(images_Arr[ind]),cv2.COLOR_GRAY2RGB)
      twoDimage = np.asarray(images_Arr[ind]).reshape((-1,3))
      twoDimage = np.float32(twoDimage)
      criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
      K = 4
      attempts=10
      ret,label,center=cv2.kmeans(twoDimage,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
      center = np.uint8(center)
      res = center[label.flatten()]
      result_image = res.reshape((np.asarray(images_Arr[ind]).shape))
      kmeans_images.append(Image.fromarray(result_image.astype('uint8'), 'RGB'))
      print('Index Done ',ind) 
    ind+=1
```

Image Segmentation using Thresholding

```python
from skimage.filters import threshold_otsu

img_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)

def filter_image(image, mask):
    r = image[:,:,0] * mask
    g = image[:,:,1] * mask
    b = image[:,:,2] * mask
    return np.dstack([r,g,b])

thresh = threshold_otsu(img_gray)
img_otsu  = img_gray < thresh
filtered = filter_image(image, img_otsu)
```
